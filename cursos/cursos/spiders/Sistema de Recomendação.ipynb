```python
import pandas as pd
import numpy as np

from math import sqrt
```

```python
def recommend_courses(user, cursos):
    courses = []
    curso_index = 0
    for curso in cursos:
        for i in range(len(user)):
            if(user[i] == curso[i] and user[i] == 1):
                courses.append(curso_index)
        curso_index += 1
    
    return pd.unique(courses)
```

```python
df = pd.read_json('cursos_udacity_10.json')
```
