```python
assuntos = ['intro', 'mobile', 'web', 'python', 'android', 'ios', 'machine learning', 'google', 'data science', 'vr']
```

```python
for assunto in assuntos:
    df[assunto] = np.zeros_like(df.headline)
```

```python
for i in range(len(df)):
    headline = df.headline[i].lower()
    title = df.title[i].lower()
    
    if('introducao' in headline or 'fundamentos' in headline or 'introducao' in title or 'fundamentos' in title):
        df['intro'][i] = 1
    elif('web' in headline or 'web' in title or 'client-server' in headline or 'client-server' in title or
        'html' in headline or 'html' in title):
        df['web'][i] = 1
    elif('google' in headline or 'google' in title):
        df['google'][i] = 1
    elif('data' in headline or 'data' in title):
        df['data science'][i] = 1
    elif('aprendizado de maquina' in headline or 'aprendizado profundo' in headline or
         'aprendizado de maquina' in title or 'aprendizado profundo' in title or
         'natural language' in headline or 'natural language' in title or
         'computer vision' in headline or 'computer vision' in title or
         'reinforcement learning' in headline or 'reinforcement learning' in title or
         'inteligencia artificial ' in headline or 'inteligencia artificial ' in title or 'AI' in headline or
         'artificial intelligence' in headline or 'artificial intelligence' in title or 'deep learning' in headline or
         'deep learning' in title or 'aprendizagem'):
        df['machine learning'][i] = 1
    elif('android' in headline or 'android' in title or 'java' in headline or 'java' in title):
        df['android'][i] = 1
    elif('mobile' in headline or 'mobile' in title):
        df['mobile'][i] = 1
    elif('ios' in headline or 'ios' in title):
        df['ios'][i] = 1
    elif('python' in headline or 'python' in title):
        df['python'][i] = 1
    elif('vr' in headline or 'vr' in title):
        df['vr'][i] = 1
```

```python
df.to_json('cursos_udacity_10.json')
```
