```python
import nltk
import pandas as pd
import re

from sklearn.naive_bayes import GaussianNB
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
from string import punctuation
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
```

```python
df = pd.read_json('cursos_udacity_8.json')
```

```python
df
```

```python
corpus = []

for i in range(len(df)):
    headline = re.sub('[^a-zA-Z]', ' ', df.headline[i])
    headline = headline.lower()
    headline = headline.split()
    ps = PorterStemmer()
    headline = [ps.stem(word) for word in headline if not word in set(stopwords.words('portuguese'))]
    headline = ' '.join(headline)
    corpus.append(headline)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
df_test = df.copy()
```

```python
for x in X:
    df_test = df_test.append(pd.Series(x), ignore_index=True)
```

```python
df_test= df_test.drop(range(74, 148))
```

```python
df_test = df_test.fillna(0)
```

```python
df_test.to_json('cursos_udacity_9.json')
```
